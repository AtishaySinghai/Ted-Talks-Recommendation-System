```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction import text
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv("ted_talks.csv")
```

```python
data.head()
```

```python
data["title"] = data["url"].map(lambda x:x.split("/")[-1])
data["title"] = data["title"].map(lambda x:x.split("\n")[0].capitalize())
data["title"] = data["title"].str.replace("_"," ")
```

```python
data
```

```python
ted_talks = data["transcript"].tolist()

uni_tfidf = text.TfidfVectorizer(input=ted_talks, norm='l2', stop_words="english")
uni_matrix = uni_tfidf.fit_transform(ted_talks)

uni_sim = cosine_similarity(uni_matrix)
```

```python
def recommend_ted_talks(x):
    return ".".join(data["title"].loc[x.argsort()[-6:-1]])
    
data["ted_talks_uni"] = [recommend_ted_talks(x) for x in uni_sim]
```

```python
mine = np.array(data['ted_talks_uni'].str.strip().str.split("\n")[1])
```

```python
result=' '.join(mine).split(".")[::-1]
```
